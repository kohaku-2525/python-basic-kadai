```python
def calculate_total(price):
  total = price + 500

  print(f"{total}円")

calculate_total(1200)
```

```python
def add_two_arguments(price, shipping_fee):
  total = price + shipping_fee

  print(f"{total}円")

add_two_arguments(1200, 500)
```

```python
# greet_user 関数の定義
def greet_user(user_name: str) -> str:
  if not user_name:
    return "こんにちは、ゲストさん！"
  else:
    return f"こんにちは、{user_name}さん！"

# 引数として '侍 太郎' を渡した場合の実行結果
print(greet_user('侍 太郎'))

# 引数として空文字を渡した場合の実行結果
print(greet_user(''))
```
