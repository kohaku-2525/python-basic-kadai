```python
stationery_prices = {"鉛筆":70,"ボールペン":120,"シャープペンシル":150,"消しゴム":90}

# ボールペンの価格を取り出して表示
print(stationery_prices["ボールペン"])
```

```python
sports = {1:"サッカー" , 2:"野球" , 3:"バスケット"}

# キー「2」の値を「バトミントン」に書き換える
sports[2] = "バトミントン"

# キー「4」として「水泳」を追加する
sports[4] = "水泳"

# 連想配列全体を取り出して表示
print(sports)
```

```python
stationery_prices = {"鉛筆" : 70, "ボールペン" : 120, "シャープペンシル" : 150,"消しゴム" : 90}

# 「シャープペンシル」を要素ごと削除
stationery_prices.pop("シャープペンシル")

# 連想配列全体を取り出して表示
print(stationery_prices)
```

```python
personal_deta = {"name":"侍太郎", "age":"36", "gender":"男性"}

# キー「name」の値を表示
print(personal_deta["name"])

# 連想配列全体を表示
print(personal_deta)

# キー「age」の値を更新
personal_deta["age"] = 37

# 新しく要素（キーと値）と追加
personal_deta["address"] = "東京"

# 連想配列全体を表示
print(personal_deta)
```
