```python
import datetime

now = datetime.datetime.now()

td_1y = datetime.timedelta(days=365)

add = now + td_1y

td_3d = datetime.timedelta(days=3)

sub = now - td_3d

print(add.strftime("現在から１年後は%Y年%m月%d日%H時%M分%S秒です。"))
print(sub.strftime("現在から３日前は%Y年%m月%d日%H時%M分%S秒です。"))
```

```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人である")
        else:
            print(f"{self.name}は大人でない")

# Humanクラスのインスタンスを複数生成
human1 = Human("侍花子", 25)
human2 = Human("侍一郎", 18)
human3 = Human("侍太郎", 30)
human4 = Human("侍三郎", 15)

# リストに追加
human_list = [human1, human2, human3, human4]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in human_list:
    human.check_adult()
```
